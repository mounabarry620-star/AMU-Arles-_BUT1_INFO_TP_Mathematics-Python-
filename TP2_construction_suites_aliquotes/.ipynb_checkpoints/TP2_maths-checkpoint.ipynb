```python
def parties_aliquotes(n):
	if n <= 1:
		return []
	diviseurs = []
	for i in range(1, n//2 + 1):
		if n % i == 0:
			diviseurs.append(i)
	return diviseurs

def suivant_aliquote(n):
	if n <= 1:
		return 0
	return sum(parties_aliquotes(n))

def parfaits(n):
	parfaits = []
	for i in range(1, n+1):
		if i == suivant_aliquote(i):
			parfaits.append(i)
	return parfaits

def amicaux(n):
	
	amicaux = []
	for a in range(2, n + 1):
		b = suivant_aliquote(a)
		if b > a and b <= n and suivant_aliquote(b) == a:
			amicaux.append((a, b))
	return amicaux

L = amicaux(1000)
print(L)
```

```python
L = amicaux(10000)
print(L)
```
