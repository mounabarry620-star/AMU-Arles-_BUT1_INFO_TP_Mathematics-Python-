```python
def parties_aliquotes(n):
     if n <= 1:
        return []                    
    
     diviseurs = []
    
     for i in range(1, int(n/2) + 1):
        if n % i == 0:
            diviseurs.append(i)
            
     return diviseurs


```

```python
n = 24
parties_aliquotes(n)
```

```python
def suivant_aliquote(n):
   if n <=1 :
     return 0

   return sum(parties_aliquotes(n))
```
