# EXERCICE 4:

```sage
# 1. Preuve:
# Si x(n) = 1 pour un certain n, alors : x(n+1) = b*x(n) − a = (a+1)*1 − a = a + 1 − a = 1
# Comme x(1) = 1, par récurrence, x(n) = 1 pour tout n
```

```sage
#2. fonction:
def f(a, n):
    b = a + 1
    x = 1
    resultats = [x]
    for i in range(1, n):
        x = b * x - a
        resultats.append(x)
    return resultats
```

```sage
# tests: a = 1023,1 et n = 10:
f(1023.1, 10)
```

```sage
# Les valeurs vont s'éloigner de 1 à cause des erreurs d'arrondi
```

```sage
# 1. Y(100) en utilisant le calcul : expression a:
a = 3.0
y_a = 0.5
for i in range(100):
    y_a = (a + 1) * y_a - a * y_a^2
print(y_a)
```

```sage
# expression b:
a = 3.0
y_b = 0.5
for i in range(100):
    y_b = (a + 1) * y_b - a * y_b * y_b
print(y_b)
```

```sage
#La seule différence : a * y_a^2 (a) vs a * y_b * y_b (b). 
# Mathématiquement c'est pareil (y^2 = y × y). Mais les résultats sont différence
```

```sage
def dif(a, n):
    y_a = 0.5
    y_b = 0.5
    differences = []
    for i in range(n):
        y_a = (a + 1) * y_a - a * y_a^2
        y_b = (a + 1) * y_b - a * y_b * y_b
        differences.append((i, y_a - y_b))
    G = list_plot(differences, plotjoined=True)
    G.show()
    return differences
```

```sage
dif(3, 100)
```

```sage
# 3 Yn(n):
def Yn(n):
    a = 3.0
    y = [0.5, 0.5, 0.5, 0.5, 0.5]
    for i in range(n):
        y[0] = (a + 1) * y[0] - a * y[0]^2           # (a)
        y[1] = (a + 1) * y[1] - a * y[1] * y[1]       # (b)
        y[2] = ((a + 1) - a * y[2]) * y[2]             # (c)
        y[3] = a * y[3] + (1 - a * y[3]) * y[3]        # (d)
        y[4] = y[4] + a * (y[4] - y[4]^2)              # (e)
    return y
```

```sage
resultats = Yn(100)
for i, val in enumerate(resultats):
    print(val)
```
