```python
def Fibo(n):
    if n <= 0: return []
    if n == 1: return [0]
    
    L = [0, 1]
    for i in range(n - 2):
        nouveau = L[-1] + L[-2]
        L.append(nouveau)
        
    return L

print(Fibo(10))
```
