# Exercice 1. Soient les matrices:

```python

var('x')

A = matrix([[1, 2, 3],
            [-1, 0, 1],
            [0, 1, 0]])

B = matrix([[2, -1, 0],
            [-1, 0, 1]])

u = matrix([[1],
            [x],
            [x^2]]) 

v = matrix([[1],
            [0],
            [1]])   

print(" Question 1 : Produits possibles ")

try:
    print("A * v =")
    print(A * v)
except:
    print("A * v impossible")

try:
    print("\nB * A =")
    print(B * A)
except:
    print("B * A impossible")

try:
    print("\nB * v =")
    print(B * v)
except:
    print("B * v impossible")
print(B * u)


print(" Question 2 : (A - I)^7 et extraction ")
I3 = identity_matrix(3)
M_res = (A - I3)^7
print("Matrice résultat :")
print(M_res)
print("Coefficient en position (2,3) [index 1,2] :", M_res[1, 2])


print("\n--- Question 3 : Inverse de A et Trace ---")
A_inv = A.inverse()
print("Inverse de A :")
print(A_inv)
print("Trace de l'inverse :", A_inv.trace())
```

# Exercice 2: Commutativité

```python


print("--- Question 1 : A et B ---")
A = matrix([[2, -1, -1, -2],
            [4, 4, 2, 0],
            [0, 4, -2, 0],
            [-1, -1, 0, -3]])

B = matrix([[-2, 2, 1, -3],
            [1, -3, 0, -2],
            [2, -1, -2, 3],
            [-4, 3, 3, 3]])


calc1 = (A + B)^2
calc2 = A^2 + 2*A*B + B^2

print("(A + B)^2 =")
print(calc1)
print("\nA^2 + 2AB + B^2 =")
print(calc2)


if calc1 == calc2:
    print("Conclusion : Les résultats sont identiques.")
else:
    print("Conclusion : Les résultats sont différents car A et B ne commutent pas (AB != BA).")


print("\n--- Question 2 : C et D ---")
C = matrix([[0, -4, 3, 1],
            [4, 2, -4, 1],
            [1, -1, 3, 3],
            [4, 1, 4, -4]])

D = matrix([[-8, -6, 26, 0],
            [4, -8, 0, -11],
            [10, -5, 26, -6],
            [-12, -23, 0, 38]])

calc3 = (C + D)^2
calc4 = C^2 + 2*C*D + D^2

if calc3 == calc4:
    print("Pour C et D, les résultats sont identiques. Cela signifie que C et D commutent (CD = DC).")
else:
    print("Les résultats sont différents.")
```

# Exercice 3: Polynôme annulateur et inverse

```python


M = matrix([[-5, -5, 2],
            [6, 8, -4],
            [6, 5, -1]])
I3 = identity_matrix(3)

print("--- Question 1 : Calcul du polynôme ---")
res = M^3 - 2*M^2 - 5*M
print("M^3 - 2M^2 - 5M =")
print(res)

print("\n--- Question 2 : Déduction de l'inverse ---")


N = (1/6) * (M^2 - 2*M - 5*I3)
print("Inverse déduit N :")
print(N)

print("\n--- Question 3 : Vérification ---")
print("Inverse calculé par Sage (M^-1) :")
print(M.inverse())

print("Est-ce que N == M.inverse() ?", N == M.inverse())
```
