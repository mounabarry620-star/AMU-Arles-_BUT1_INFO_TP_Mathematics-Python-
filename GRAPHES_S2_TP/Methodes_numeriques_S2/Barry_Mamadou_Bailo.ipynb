# Exercice 1 : Algorithme de Horner

```sage
P = [-3, 5, 2, 7, -2, 3] 
```

```sage
def Poly(P, alpha):
    resultat = 0
    n = len(P)
    for i in range(n):
        resultat = resultat + P[i] * alpha^i
    return resultat


%time Poly(P, 1)
```

```sage
def Horner(P, alpha):
    n = len(P) - 1
    A = P[n] 
    for j in range(n - 1, -1, -1):
        A = P[j] + alpha * A
    return A
    
%time Horner(P, 1)

```

# Exercice 2 : Dérivation 


```sage
def deriv(f, x, h=0.001):
    return (f(x + h) - f(x)) / h
```

```sage
deriv(ln, 2.)
```

# Exercice 3 : Dérivation symétrique (centrée)



```sage
def deriv2(f, x, h=0.001):
    return (f(x + h/2) - f(x - h/2)) / h
```

```sage
h = 10**(-8)
deriv2(ln, 2., h)
```

```sage
f(x)= 1/x
deriv2(f, 2., h)
```

# Exercice 4 : Approximation par Taylor d'ordre 1


```sage
f(x) = 5/x
x0 = 2
P1(x) = f(x0) + diff(f, x)(x0) * (x - x0)
print("P1(x) =", P1(x))

# Graphique de f et P1 sur ]1/4, 5]
plot(f, (x, 1/4, 5), color='blue') + \
plot(P1, (x, 1/4, 5), color='red')
```

```sage
# Graphique de l'erreur e1(x) = f(x) - P1(x) sur ]1, 3]
e1(x) = f(x) - P1(x)
plot(e1, (x, 1, 3), color='green', title="Erreur d'ordre 1 : e1(x)")
```

# Exercice 5 : Dérivation seconde numérique


```sage
def deriv_seconde(f, x, h=0.01):
    return (f(x + h) - 2*f(x) + f(x - h)) / h^2

deriv_seconde(ln, 2.)
```

# Exercice 6 : Taylor d'ordre 2


```sage
# P2(x)
P2(x) = P1(x) + (diff(f, x, 2)(x0) / 2) * (x - x0)^2
print("P2(x) =", P2(x))

plot(f, (x, 1/4, 5), color='blue') + \
plot(P1, (x, 1/4, 5), color='red') + \
plot(P2, (x, 1/4, 5), color='purple')
```

```sage
e2(x) = f(x) - P2(x)
plot(e1, (x, 1, 3), color='green') + \
plot(e2, (x, 1, 3), color='orange')
```

# Exercice 7 : Approximation du sinus par série de Taylor


```sage
def sinus(x, n):
    somme = 0
    for k in range(n + 1):
        
        terme = ((-1)^k * x^(2*k + 1)) / factorial(2*k + 1)
        somme = somme + terme
    return somme

# Test avec x = pi/3 et n = 10
x = pi/3
print("Approximation de sin(pi/3) avec n=10 :", sinus(x, 10).n())
print("Valeur exacte sin(pi/3) :", sin(x).n())
```
